```python
def writeBase64Image(imagestr,filename) :
    print(filename)
    image = open(filename, "wb") # open the file with filename as its name, enable writing and specify to handle it in binary mode
    imgdata = base64.b64decode(str(imagestr)) # take the image-string from the response and decode it from base64 to png
    print(type(imgdata))
    image.write(imgdata)
    image.close()
```
